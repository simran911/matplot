Matplotlib.

Matplotlib is a popular Python library for data visualization. It provides a wide range of functions and tools for creating high-quality plots, charts, and figures. Matplotlib is widely used in various domains such as data analysis, scientific research, and data visualization tasks.

Matplotlib is used for:

Data exploration and analysis: Matplotlib allows you to visualize data in different formats, making it easier to understand patterns, trends, and relationships within the data.

Presentation and communication: Matplotlib enables the creation of visually appealing plots and charts that can effectively communicate insights and findings to an audience.

Report generation: Matplotlib can be used to generate plots and figures for inclusion in reports, articles, or scientific papers.

GUI development: Matplotlib can be integrated into graphical user interfaces (GUIs) to provide interactive and dynamic data visualization capabilities.

Educational purposes: Matplotlib is widely used for teaching and learning data visualization concepts due to its simplicity and flexibility.

Q2: Name five plots that can be plotted using the Pyplot module of Matplotlib.

The Pyplot module in Matplotlib provides a simple interface for creating various types of plots. Here are five commonly used plots that can be plotted using the Pyplot module:

Line Plot: A line plot displays data points connected by straight lines. It is useful for visualizing trends and changes over continuous variables.

Bar Plot: A bar plot represents data using rectangular bars of varying heights. It is suitable for comparing categorical data or showing frequency counts.

Scatter Plot: A scatter plot shows the relationship between two variables as a collection of points. It helps in understanding the correlation or patterns between variables.

Histogram: A histogram is used to visualize the distribution of a single variable by dividing it into bins and displaying the frequency or density of data within each bin.

Pie Chart: A pie chart represents data as a circle divided into sectors, where each sector represents a category or proportion of a whole. It is effective for displaying categorical data as percentages or proportions.

These are just a few examples, and Matplotlib offers many more plot types and customization options to cater to different data visualization needs.

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([0, 1, 2, 3, 4, 5])

# Line 1
y1 = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(2, 2, 1)
plt.plot(x, y1)
plt.title('Line 1')

# Line 2
y2 = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(2, 2, 2)
plt.plot(x, y2)
plt.title('Line 2')

# Line 3
y3 = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(2, 2, 3)
plt.plot(x, y3)
plt.title('Line 3')

# Line 4
y4 = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(2, 2, 4)
plt.plot(x, y4)
plt.title('Line 4')

# Adjust spacing between subplots
plt.tight_layout()

# Display the figure
plt.show()
```
